```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

```python
crash = sns.load_dataset('car_crashes')
```

```python
sns.distplot(crash['not_distracted'],kde=False)
```

```python
sns.pairplot(crash)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(tips,hue='sex',palette='Blues')
```

```python
sns.set_style('dark')
plt.figure(figsize=(8,4))
sns.jointplot(x='speeding',y='alcohol',data=crash,kind='reg')
sns.jointplot(x='speeding',y='alcohol',data=crash)
```
