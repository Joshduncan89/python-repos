```python
import pandas as pd

df = pd.read_csv('Minimum Wage Data.csv', encoding='latin')
min_wage = pd.DataFrame()
df.columns
```

```python
for name,group in df.groupby('State'):
    if min_wage.empty:
        min_wage = group.set_index('Year')[['State.Minimum.Wage']].rename(columns={'State.Minimum.Wage': name})
    else:
        min_wage = min_wage.join(group.set_index('Year')[['State.Minimum.Wage']].rename(columns={'State.Minimum.Wage': name}))
min_wage.head()
```
